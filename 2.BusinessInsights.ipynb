### Unpacking Restaurant Trends
### LEVEL 2: Table Booking, Price Range Analysis, Feature Engineering
We’ll dive deeper into how restaurants operate—Do they offer table booking? How does pricing affect ratings? Which features make a restaurant more popular? We’ll extract new insights and create useful features to better understand restaurant trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset.csv")
```

```python
print("Table Booking %:\n", df['Has Table booking'].value_counts(normalize=True) * 100)
print("Online Delivery %:\n", df['Has Online delivery'].value_counts(normalize=True) * 100)
```

```python
df.groupby('Has Table booking')['Aggregate rating'].mean().plot(kind='bar', title="Table Booking vs Rating")
plt.show()
```

```python
df.groupby('Price range')['Has Online delivery'].value_counts().unstack().plot(kind='bar', title="Price Range vs Online Delivery")
plt.show()
```

```python
print("\nMost Common Price Range:\n", df['Price range'].value_counts())
```

```python
df.groupby('Price range')['Aggregate rating'].mean().plot(kind='bar', title="Price Range vs Rating")
plt.show()
```

```python
df['Restaurant_Name_Length'] = df['Restaurant Name'].apply(len)
df['Has_Table_Booking'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})
df['Has_Online_Delivery'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})
print("\nNew Features Created:\n", df[['Restaurant_Name_Length', 'Has_Table_Booking', 'Has_Online_Delivery']].head())
```

Conclusion:
By analyzing booking systems, pricing, and customer preferences, we’ve uncovered key trends in the restaurant industry. We now know how pricing affects ratings, which cities have the best dining experiences, and how features like online delivery impact popularity. These insights set the stage for predictive modeling!
